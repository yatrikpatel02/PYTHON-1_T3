```python
#wa pager program your solution should prompt for file name and display the text file 5 lines at a time causing each time to ask the user to press any key to continue and press q to quit
file=input("Enter file name : ")
f=open(file,'r')
s=f.readlines()
i=1
n=6
c=0
l=1

for i in range(i,n):
    print(l,' : ',s[c])
    if len(s)==i:
        break
    c+=1
    l+=1
b=True
while b!=False:
    if len(s)==l:
        break
    ch=input("Enter any character for continue or 'q' for exit : ")
    if ch=='Q' or ch=='q':
        b=False
    else:
        if c==len(s)-1:
            b=False
            print("File is ended!")
            break
        for i in range(5):
            print(l,' : ',s[c])
            c+=1
            l+=1
            if c==len(s):
                b=False
                print("File is ended!")
                break
                
```

```python
f=open("a1.txt")
f1=open("output.txt",'w')
while True:
    b=f.readline()
    if len(b)!=0:
        if b[0]=="#":
            continue
        else:
            if '#' in b:
                for index in range(1,len(b)):
                    if b[index]=="#":
                        f1.write(b[0:index]+"\n")
            else:
                f1.write(b)
    else:
        break
f.close()
f1.close()
```
