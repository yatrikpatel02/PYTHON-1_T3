```python
%%writefile calculator.py
def add(a,b):
    print("addition: ",a+b)
def sub(a,b):
    print("subtraction: ",a-b)
def multi(a,b):
    print("multiplication: ",a*b)
def div(a,b):
    print("division: ",a/b)
def power(a,b):
    print("power: ",a**b)
def modulo(a,b):
    print("modulo: ",a%b)
def floor(a,b):
    print("floor division: ",a//b)
    
```

```python
%%writefile func.py
def stringfunc(s):
    print("length: ",len(s))
    print("upper: ",s.upper())
    print("lower: ",s.lower())
    print("title: ",s.title())
    print("capitalize: ",s.capitalize())
    print("alphabet check: ",s.isalpha())
    print("alphabet and numeric check: ",s.isalnum())
    print("title check: ",s.istitle())
    print("digit check: ",s.isdigit())
def listfunc(l):
    print("length: ",len(l))
    l.append(12)
    print("append: ",l)
    l1=l.copy()
    print("copy: ",l1)
    l2=[1,2,3,4,5]
    l1.extend(l2)
    print("extend: ",l1)
```

```python
import func
val=eval(input("Enter value: "))
if type(val)==str:
    func.stringfunc(val)
elif type(val)==list:
    func.listfunc(val)
else:
    print("Enter valid data!")
```
