# writing the files
- 1) write() - write content if file without adding any extra character
- 2) writelines() - write the content of a list to the file

```python
f=open("demo.txt",'r')
f1=open("tk2.txt",'w')
s=f.readlines()
s.reverse()
f1.writelines(s)
f1.close()  
```

```python
n=int(input("Enter total no of entries: "))
l=[]
for i in range(n):
    l.append(input("Enter name: ")+"\n")
f=open("tk3.txt",'w')
f.writelines(l)
f.close()
```
