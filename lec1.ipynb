### File Access modules
- Types of operation possible in opened file
- how the file will be used once its opened
- modes also define the location of the file handle in the file
- file handle is like a cursor, which defines from where data has to be read or written in the file
- six access modes(r,r+,w,w+,a,a+)
**1) read only(r):**
- open for reading
- handle is positioned at the beggining of the file
- if file does not exits, raises the error
- this is also a default mode in which file is opened

```python
f=open("test.txt",'r')
print(f.read())
print(f.tell())
f.seek(0)                   # redirect cursor index to 0
print(f.read())
f.close()
```

### 2) Read and write('r+')
- open text file from reading and writing.
- handle is positioned at the beginning of the file
- if file does not exits, raises an error

```python
f=open("test.txt",'r+')
f.read()
f.write(" hiii")
f.seek(0)
print(f.read())
f.close()
```

### 3) write only('w')
- open text file for writing
- handle is positioned at the beginning of the file
- for existing file, the data is truncated and overwritten
- creates the file if file does not exits

##### how to open a file
- 1) f=open(file_name,access_name) - compulsory use close() after write()
- 2) with open(file_name,access_name) - no need to use close after write()

```python
f=open("demo.txt",'w')
f.write("python-1\n")
f.write("pyhton-2\n")
f.write("pyhton-3\n")
f.write("pyhton-4\n")
f.close()
```

```python
with open("new.txt","w") as f:
    f.write("python-1\n")
    f.write("pyhton-2\n")
    f.write("pyhton-3\n")
    print(f.closed)
print(f.closed)    
```

#### 4) write and read ('w+')
- open text file from writing and reading
- handle is positioned at the beginning of the file
- for existing file, the data is truncated and overwritten
- create the file if doesn't exist
- compulsory use close() after using write()

```python
f=open("demo.txt","w+")
f.write("New Data\n")
f.write("LJIET")
f.seek(0)
print(f.read())
f.close()
```

#### 5) append mode ('a')
- open for writing
- handle is positioned at the end of the file
- the data being written will be inserted at the end of the file, after the existing data
- creates the file if file does not exists

```python
f=open("demo.txt",'a')
f.write("New Data1\n")
f.write("LJIET\n")
f.write("LJENG\n")
f.write("LJU")
f.close()
```

#### 6) append mode ('a+')
- open for writing and reading
- handle is positioned at the end of the file
- the data being written will be inserted at the end of the file, after the existing data
- creates the file if file does not exists

```python
f=open("demo.txt",'a+')
f.write("New Data1\n")
f.write("LJIET\n")
f.seek(0)
print(f.read())
f.close()
```

## reading the files
- 1) read() - returns read byted in the form of string
- 2) readline()  - returns a line of the file and returns in the form of string
- 3) readlines() - reads all the line and returns them as a string element in a list

```python
f=open("demo.txt")
print(f.read())
f.close()
```

```python
f=open("demo.txt")
print(f.read(10))
print(f.read(14))
print(f.read(12))
```

```python
f=open("demo.txt")
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
f.close()
```

```python
f=open("demo.txt")
print(f.readlines(10))
print(f.readlines())
f.close()
```
