# unit - 8) oops and exception handling
- try --> block let u test a block of code for errors
- except --> block let u handle the errors 
- else--> lets u execute code when there is no error
- finally --> lets u execute code regardless of result of try and except block

```python
def div(a,b):
    return a/b
try:
    print(div(5,0))
except:
    print("Exception occured")
else:
    print("no exception")
finally:
    print("Always printed")
```
