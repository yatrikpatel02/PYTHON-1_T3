# custom Exception

```python
class tejas(Exception):
    def __init__(self,arg):
        self.msg=arg
```

```python
def add(a,b):
    if type(a)==str:
        raise tejas("personalized error message")
    if type(b)==str:
        raise tejas("personalized error message")
add(2,'a')        
```

```python
class withdrawError(Exception):
    def __init__(self,arg):
        self.msg=arg
```

```python
def withdraw(amt):
    total=10000
    try:        
        if amt>total:
            raise withdrawError("Required Amount is greater than total Amount")
        else:
            total=total-amt
            print("withdraw : ",amt)
            print("Total left after withdraw : ",total)
    except withdrawError as e:
        print(e)
withdraw(1000)     
```
